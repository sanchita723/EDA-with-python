```python
df_country = pd.read_excel("C:/Users/Lenovo/Downloads/Zomatodataset/Country-Code.xlsx")
df_country.head()
```

```python
final_df = pd.merge(df,df_country, on = "Country Code", how = "left")
final_df.head(3)
```

```python
#Top 3 countries
country_names =  final_df.Country.value_counts().index
country_names
```

```python
country_val = final_df.Country.value_counts().values
country_val
```

```python
plt.pie(country_val[:3],labels=country_names[:3], autopct= "%1.2f%%", explode = [0.2,0,0])
```

Observation: Zomato's Maximun records are form India 
after tha US then Uk

```python
#Ratings
ratings = final_df.groupby(["Aggregate rating", "Rating color", "Rating text"]).size().reset_index().rename(columns= {0:'Rating_count'})
ratings
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x= 'Aggregate rating', y = "Rating_count", data = ratings )
```

Observation : Most of the user did not give any ratings and the highest ratings are in between 3 to 4

```python
sns.countplot(x = 'Rating color', data = ratings)
```
