```python
#Adding extra column order month abd order year
df['Order date year'] = df['Order_Date'].dt.year
df['Order date month'] = df['Order_Date'].dt.month
df.head(2)
```

```python
#Profit margin
df['Profit_Margin'] = df['Profit'] / df['Sales']
df.head(2)
```

```python
# Total profit and sale by subcategory
df_by_sub_category = df.groupby('Sub-Category').sum()
df_by_sub_category.reset_index(inplace=True)
df_by_sub_category.head()
```

```python
#Creating function of groupby
def grouped_data(column_name):
    df_tmp = df.groupby(column_name).sum()
    df_tmp.reset_index(inplace=True)
    return df_tmp

#Group dataframe by ship mode
grouped_data('Ship_Mode')
```

```python
#Visulizations
#histogram
fig = px.histogram(df, x='Sales', template = 'plotly_dark')
fig.show()
```

```python
#boxplot
fig = px.box(df, y='Sales', range_y=[0,1000], template = 'plotly_dark')
fig.show()
```

```python
#Plotting sales by sub category
data = grouped_data('Sub-Category')

#barchart
fig = px.bar(data, x ='Sub-Category',
             y = 'Sales', 
             title = '<b> Sales By Sub Catrgory <b>',
             template = 'plotly_dark')
fig.show()
```

```python
#Profit by sub category
fig = px.bar(data, x ='Sub-Category',
             y = 'Profit', 
             title = '<b> Profit By Sub Catrgory <b>',
             template = 'plotly_dark')
fig.show()
```

```python
#Ploting sales and profit by sub category
fig = px.bar(data, x ='Sub-Category',
             y = 'Sales', 
             color = 'Profit',
             color_continuous_scale = ['red', 'yellow', 'green'],
             title = '<b> Sales And Profit By Sub Catrgory <b>',
             template = 'plotly_dark')
fig.show()
```

```python
#Inspecting negative profit of tables
fig = px.scatter(df, x ='Sales',
             y = 'Profit', 
             color = 'Discount',
             title = '<b> Scatterplot Sales And Profit <b>',
             template = 'plotly_dark')
fig.show()
```

```python
#Avg discount and overall profit on all subcategories
df_discount = df.groupby('Sub-Category').agg({'Discount' : 'mean',
                                              'Profit': 'sum'})
df_discount
```

```python
# creating chart
fig = px.bar(df_discount, 
             x =df_discount.index,
             y = 'Discount', 
             color = 'Profit',
             color_continuous_scale = ['red', 'yellow', 'green'],
             title = '<b> Mean Discount By Sub Catrgory <b>',
             template = 'plotly_dark')

fig.show()
```

```python
#Sorting values by date
df_sorted = df.sort_values(by=['Order_Date'])

#cumulative sales and profit
#cumulative display the total sum of data as it grows with time (or any other series or progression)

df_sorted['cumulative_sales'] = df_sorted['Sales'].cumsum()
df_sorted['cumulative_profit'] = df_sorted['Profit'].cumsum()

df_sorted.head()
```

```python
#linechart
fig = px.line(df_sorted,
             x ='Order_Date',
             y ='cumulative_sales',
             title = '<b> Cumulative Sales And Profit <b>',
             template = 'plotly_dark')

fig.show()
```

```python
fig = px.line(df_sorted,
             x ='Order_Date',
             y ='cumulative_profit',
             title = '<b> Cumulative Sales And Profit <b>',
             template = 'plotly_dark')

fig.show()
```
